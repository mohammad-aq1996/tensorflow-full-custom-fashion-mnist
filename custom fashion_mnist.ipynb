```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Flatten, Dense, Input, Layer
from tensorflow.keras.models import Sequential, Model
import tensorflow_datasets as tfds
import numpy as np
```

```python
train = tfds.load('fashion_mnist', split='train')
test = tfds.load('fashion_mnist', split='test')
```

```python
def preprocessing(data):
    image = data['image']
    image = tf.reshape(image, [-1])
    image = tf.cast(image, tf.float32)
    image = image / 255.
    return image, data['label']
```

```python
train_ds = train.map(preprocessing)
test_ds = test.map(preprocessing)
```

```python
BUFFER_SIZE = 1024
BATCH_SIZE = 64
train_dataset = train_ds.shuffle(BUFFER_SIZE).batch(BATCH_SIZE)
test_dataset = test_ds.batch(BATCH_SIZE)
```

```python
class SimpleDense(Layer):
    def __init__(self, units, activation=None):
        super(SimpleDense, self).__init__()
        self.units = units
        self.activation = keras.activations.get(activation)

    def build(self, input_shape):
        w_init = tf.random_normal_initializer()
        self.w = tf.Variable(name='kernel', initial_value=w_init(shape=(input_shape[-1], self.units), dtype=tf.float32), trainable=True)
        b_init = tf.zeros_initializer()
        self.b = tf.Variable(name='bias', initial_value=b_init(shape=(self.units,), dtype=tf.float32), trainable=True)

    def call(self, inputs):
        return self.activation(tf.matmul(inputs, self.w)+self.b)
```

```python
class SimpleModel(Model):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.hidden = SimpleDense(128, 'relu')
        self.out = SimpleDense(10, 'softmax')

    def call(self, inp_shape):
        inp = Input(shape=inp_shape)
        x = self.hidden(inp)
        out = self.out(x)
        model = Model(inp, out)
        return model
```

```python
model = SimpleModel()
model=model((784,))

model.summary()
```

```python
my_loss = keras.losses.SparseCategoricalCrossentropy()
optimizer = keras.optimizers.Adam()
train_acc_metric = keras.metrics.SparseCategoricalAccuracy()
val_acc_metric = keras.metrics.SparseCategoricalAccuracy()
```

```python
def train_for_one_epoch():
    losses=[]
    for (x_batch_train, y_batch_train) in train_dataset:
        logits, loss = apply_grads(optimizer, model, x_batch_train, y_batch_train)
        losses.append(loss)
        train_acc_metric(y_batch_train, logits)
    return losses
```

```python
def perform_validation():
    losses = []
    for x_batch_val, y_batch_val in test_dataset:
        logits = model(x_batch_val)
        loss = my_loss(y_batch_val, logits)
        losses.append(loss)
        val_acc_metric(y_batch_val, logits)
    return losses
```

```python
@tf.function
def apply_grads(optimizer, model, x, y):
    with tf.GradientTape() as tape:
        logits = model(x)
        loss = my_loss(y, logits)
    grads = tape.gradient(loss, model.trainable_weights)
    optimizer.apply_gradients(zip(grads, model.trainable_weights))
    return logits, loss
```

```python
epochs = 4
epoch_train_loss, epoch_val_loss = [], []
epoch_train_acc, epoch_val_acc = [], []
for epoch in range(epochs):
    print(f'Start of epoch: {epoch}')
    losses_train = train_for_one_epoch()
    train_acc = train_acc_metric.result()

    losses_val = perform_validation()
    val_acc = val_acc_metric.result()

    losses_train_mean = np.mean(losses_train)
    losses_val_mean = np.mean(losses_val)

    epoch_train_loss.append(losses_train_mean)
    epoch_val_loss.append(losses_val_mean)

    epoch_train_acc.append(train_acc)
    epoch_val_acc.append(val_acc)

    print(f'\n Epoch {epoch:2d}: train loss: {losses_train_mean:.4f}, train accuracy: {train_acc:.4f}, validation loss: {losses_val_mean:.4f}, validation accuracy: {val_acc:.4f}')

    train_acc_metric.reset_state()
    val_acc_metric.reset_state()
```

```python
import matplotlib.pyplot as plt

plt.plot(epoch_train_acc)
plt.plot(epoch_val_acc)
plt.title('Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.show()
```
